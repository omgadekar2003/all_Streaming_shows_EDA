```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualize numerical data distributions using histograms
sns.histplot(data['IMDB Rating'], bins=20, kde=True)
plt.title('Distribution of Numerical Column')
plt.show()
```
